```python
import speech_recognition as sr
```

```python
list_of_function_in_sr= dir(sr)
print(list_of_function_in_sr)
```

```python
r = sr.Recognizer()
with sr.AudioFile(audio_file_name) as source:
    audio_file = r.record(source)  # read the entire audio file
    text_file = r.recognize_google(audio_file)
    print(text_file)
```

```python
with open('audio_to_text_output.txt','w') as f:
  f.writelines(text_file)
```
